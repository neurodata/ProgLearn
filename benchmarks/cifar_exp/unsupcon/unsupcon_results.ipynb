```python
import matplotlib as mpl

mpl.rcParams["axes.formatter.useoffset"] = False
import matplotlib.pyplot as plt
import numpy as np
import pickle
import tensorflow as tf
```

```python
with open(r"loss_batch.pickle", "rb") as infile:
    loss_batch = pickle.load(infile)
with open(r"loss_epoch.pickle", "rb") as infile:
    loss_epoch = pickle.load(infile)
```

```python
plt.figure()
plt.plot(np.arange(len(loss_batch)), loss_batch)
plt.title("Batch Loss")
plt.xlabel("batch num")
plt.ylabel("unsupervised contrastive loss")
plt.figure()
plt.plot(np.arange(len(loss_epoch)), loss_epoch)
plt.title("Epoch Average Loss")
plt.xlabel("epoch num")
plt.ylabel("avg unsupervised contrastive loss");
```
